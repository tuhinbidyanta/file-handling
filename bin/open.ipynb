```python
import struct
import pandas as pd

# Define the structure of the binary data
# Example: 2 integers followed by a float (adjust according to your file's structure)
record_format = 'iif'

# Calculate the size of each record
record_size = struct.calcsize(record_format)

# Read the binary file
data = []
with open('indiana.bin', 'rb') as file:
    while chunk := file.read(record_size):
        record = struct.unpack(record_format, chunk)
        data.append(record)

# Convert the data into a DataFrame
df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3'])

df
```
